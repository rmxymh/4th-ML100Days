## [作業目標]
了解機器學習適合應用的領域與範疇。

## [作業重點]
透過瞭解機器學習的運作方式，找出非常適用機器學習應用的領域及其原因

## 作業
請找出一個你認為機器學習可以發光發熱的領域並闡述原因? 

### Answer ###

我認為機器學習在很多領域中可以發光發熱，在這個部分可以涉及物品辨識（可應用在社群網站或app審核上）、更進一步的也可以應用在農作品改良上。而這次主要是想針對『農作品改良』的部分來說明。

之前有看過日本的節目有針對『蘋果』來進行檢驗，一般來說會是品種偏甜、顆粒較大的會比較有賣相，銷售的成果也會更好。而在這點來說，我在那個節目中看到在採收後，檢驗人員運用機器下去對於大小、外觀下去進行分類，這部分會初步將採收的蘋果分級。而這部分也是在機器學習上的辨識與分類所做出來的應用，也能夠結合工業自動化的方式來增進效率。

另外更進一步的，會針對一些良品進行檢驗，方法主要是透過針頭取得部份的蘋果汁，並且試圖迅速分析成份來判斷這些蘋果的口味如何，來進行更細的行銷規劃。在這個部分在量大的時候也可以試著透過supervised learning下去做model training，並且搭配多個成分來進行維度分析，使得能夠進一步分析出這些蘋果到底口味偏酸、偏甜、對特殊環境的抵抗能力如何，甚至是適合拿來做什麼行銷應用。這部分一方面可以搭配做行銷，另一方面則可應用在蘋果品種改良的實驗上。

也由於品種越來越多，產能的部分也會增加，這部分如果以一般人力來說反而會花很多時間，也很難在短時間之內對品質有同樣的規範，因此我自己認為在這段上，農業、食品等領域，搭配機器學習以及工業自動化的方式，能夠有效節省能力並且增進檢驗與日後品種改良的效率。
